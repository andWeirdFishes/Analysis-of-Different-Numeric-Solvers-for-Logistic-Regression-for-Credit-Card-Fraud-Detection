**Let's plot the distributions of the instances against their classes next to see if we notice any features in which the difference between distributions of the same feature in differing classes is significant:**

```python
features = df.iloc[:,:-1].columns
cls = df.columns[-1]

sns.set(style="whitegrid")

for feature in features:
    plt.figure(figsize=(12, 6))
    sns.histplot(data=df[df[cls] == 0], x=feature, bins=30, kde=False, color='green', label='Class 0', element='step', stat='probability')
    sns.histplot(data=df[df[cls] == 1], x=feature, bins=30, kde=False, color='red', label='Class 1', element='step', stat='probability')
    plt.title(f'Distribution of {feature} by Class')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend(title='Class')
    plt.tight_layout()
    plt.show()
```

We notice that the conditional distributions of certain features, such as V12, differ significantly across the two classes, suggesting a strong class-dependent relationship. This will affect our decision in the feature selection when it comes to it

```python
import pandas as pd

outlier_summary = []

features = df.iloc[:, :-1].columns
cls = df.iloc[:, -1]

for feature in features:
    data = df[feature]
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outlier_mask = (data < lower_bound) | (data > upper_bound)
    total_outliers = outlier_mask.sum()

    outliers_class0 = outlier_mask & (cls == 0)
    outliers_class1 = outlier_mask & (cls == 1)

    count_class0 = outliers_class0.sum()
    count_class1 = outliers_class1.sum()

    outlier_summary.append({
        "feature_name": feature,
        "outlier_count": total_outliers,
        "outlier_count_class0": count_class0,
        "outlier_count_class1": count_class1
    })

outlier_df = pd.DataFrame(outlier_summary)

outlier_df
```

We can't rely on dropping outliers since:

1. If we decide to drop all rows containing outliers, almost all of class 1 will be gone, as some features have 400+ outliers that are classified as class 1;
2. If we decide to drop only class 0 rows containing outliers, we will teach the model that all rows with outlier values belong to class 1 which may generate a lot of false positives on new data

Hence, this will affect our choice of scaling. We know that standardizing the data can diminish the effect of outliers, so we will use ROBUST SCALING!

```python
X = df.drop(columns='Class')
y = df['Class']
print(f'Shape of feature table: {X.shape}; Shape of target variable: {y.shape}')
```

# Training models

In each subsection (except the hyperparameter tuning subsection) we will train 5 models differing in the numeric solver we use

## On non-undersampled data

```python
from sklearn.linear_model import LogisticRegression
from logistic import CustomLogisticRegression

newton = LogisticRegression(solver='newton-cholesky')
fh = CustomLogisticRegression(solver='fixed-hessian')
cg = LogisticRegression(solver='newton-cg')
bfgs = CustomLogisticRegression(solver='bfgs')
lbfgs = LogisticRegression()
```

```python
models = [newton, fh, cg, bfgs, lbfgs]
import time

train_times = {}

for model in models:
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()
    train_times[model.solver] = end - start
```

```python
# Making predictions on our test data
y_newton, y_fh, y_cg, y_bfgs, y_lbfgs = [model.predict(X_test) for model in models]
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report

def plot_cm(y_true, y_pred, title="Confusion Matrix", labels=None, cmap='plasma'):
    disp = ConfusionMatrixDisplay.from_predictions(
        y_true, y_pred, display_labels=labels, cmap=cmap
    )
    disp.ax_.set_title(title)
    plt.show()
```

```python
y_preds = [y_newton, y_fh, y_cg, y_bfgs, y_lbfgs]

for y_pred, model in zip(y_preds, models):
    print(f"=== Classification Report for {model.solver} ===")
    print(classification_report(y_test, y_pred))
    print(f"*** Time taken to train {model.solver}: {train_times[model.solver]}s ***")
    plot_cm(y_test, y_pred, title=f"Confusion Matrix: {model.solver}")
    print("\n")
```

We notice that the "fancy algorithms" (BFGS/LBFGS) have produced only ONE LESS false positive (in a dataset of 250000 instances), therefore improving the precision ever-so-slightly but even the simple Newton-Cholesky algorithm which took only 0.8 seconds to train performs really well. However, we expected the recall to be bad here because of the huge imbalance. In the next subsection, we will undersample the 0 class in the train set and see how that holds up.

## Undersampling

We will undersample the training data to balance the classes. A perfect balance may be too extreme and lead to a lot of false positives. That's why we do a 75-25 sampling strategy.

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy={0: 1032, 1: 344}, random_state=2)
Xu_train, yu_train = rus.fit_resample(X_train, y_train)
```

```python
u_train_times = {}

# Fitting new data in the same models

for model in models:
    start = time.time()
    model.fit(Xu_train, yu_train)
    end = time.time()
    u_train_times[model.solver] = end - start
```

```python
yu_newton, yu_fh, yu_cg, yu_bfgs, yu_lbfgs = [model.predict(X_test) for model in models]
yu_preds = [yu_newton, yu_fh, yu_cg, yu_bfgs, yu_lbfgs]
```

```python
for y_pred, model in zip(yu_preds, models):
    print(f"=== Classification Report for {model.solver} ===")
    print(classification_report(y_test, y_pred))
    print(f"*** Time taken to train {model.solver}: {train_times[model.solver]}s ***")
    plot_cm(y_test, y_pred, title=f"Confusion Matrix: {model.solver}")
    print("\n")
```

It is worth noting that Fixed-Hessian being the simplest numerical method actually has the highest F1-score and the least amount of false positives.

## Undersampling & automatic feature selection

In this subsection, we will use statistical filter methods  and wrapper methods to determine the optimal subset of features. The dataset comes with anonymized features which are the result of PCA, so we do not need to reduce dimensionality with similar techniques again. We will use a train-validation-test split for the following section, so that we are not biased in our selection.

```python
X_val, X_test_new, y_val, y_test_new = train_test_split(X_test, y_test, test_size=0.5, random_state=69, stratify=y_test)
```

The models produce very similar outcomes and they differ in less than 1% of the predictions, so we do not need to train a different recursive feature eliminator for each model, as it will probably generate the same subset of features anyways.

```python
from sklearn.feature_selection import RFE
from sklearn.metrics import classification_report
import numpy as np

for i in range(10, 30):
    rfe = RFE(estimator=LogisticRegression(), n_features_to_select=i)
    rfe.fit(Xu_train, yu_train)
    selected_features = np.where(rfe.support_)[0]
    X_val_selected = X_val.iloc[:, selected_features]
    y_pred = rfe.estimator_.predict(X_val_selected)
    print(f"\n--- Classification Report with {i} features ---")
    print(classification_report(y_val, y_pred, digits=6))
```

We don't want to sacrifice the recall, but we also want to improve the precision. The subset of 21 features is satisfactory since the F1 score achieves a local maximum for n_features=21.

```python
rfe = RFE(estimator=LogisticRegression(), n_features_to_select=21)
rfe.fit(Xu_train, yu_train)
selected_features = np.where(rfe.support_)[0]
Xf_val = X_val.iloc[:, selected_features]
Xf_test_new = X_test_new.iloc[:, selected_features]
Xf_train = Xu_train.iloc[:, selected_features]
```

```python
f_train_times = {}

# Fitting new data in the same models

for model in models:
    start = time.time()
    model.fit(Xf_train, yu_train)
    end = time.time()
    f_train_times[model.solver] = end - start
```

```python
yf_newton, yf_fh, yf_cg, yf_bfgs, yf_lbfgs = [model.predict(Xf_test_new) for model in models]
yf_preds = [yf_newton, yf_fh, yf_cg, yf_bfgs, yf_lbfgs]

for y_pred, model in zip(yf_preds, models):
    print(f"=== Classification Report for {model.solver} ===")
    print(classification_report(y_test_new, y_pred))
    print(f"*** Time taken to train {model.solver}: {train_times[model.solver]}s ***")
    plot_cm(y_test_new, y_pred, title=f"Confusion Matrix: {model.solver}")
    print("\n")
```

We notice that the fixed-hessian method tends to sacrifice a little recall in favor of severe reduction of the false positive rate. Choosing whether or not to implement this numeric solver depends on how much more we want to prioritize maximizing the recall.

**This choice of features and split gave us the best results so far, so this is the model we will use in our hyperparameter tuning subsection.**

## Undersampling & Hyperparameter tuning using the Optuna framework

```python
import optuna
from sklearn.metrics import recall_score, precision_score, f1_score, roc_auc_score, average_precision_score

sklearn_solvers = ['lbfgs', 'newton-cholesky', 'newton-cg']
custom_solvers = ['bfgs', 'fixed-hessian']
all_solvers = sklearn_solvers + custom_solvers
best_models = {}
best_scores = {}
def custom_fraud_score(y_true, y_pred, y_prob=None):
    recall = recall_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    f_beta = (1 + 2**2) * (precision * recall) / ((2**2 * precision) + recall) if (precision + recall) > 0 else 0
    pr_auc = average_precision_score(y_true, y_prob) if y_prob is not None else 0
    score = 0.7*precision + 0.3*f_beta

    return score, recall, precision, f_beta, pr_auc if y_prob is not None else None

for solver in all_solvers:
    def objective(trial):
        common_params = {
            'C': trial.suggest_float('C', 1e-4, 1e4, log=True),
            'tol': trial.suggest_float('tol', 1e-6, 1e-2, log=True),
            'max_iter': trial.suggest_int('max_iter', 100, 1000),
            'random_state': 42
        }
        class_weight_option = trial.suggest_categorical('class_weight_option', ['balanced', 'custom', None])
        if class_weight_option == 'balanced':
            common_params['class_weight'] = 'balanced'
        elif class_weight_option == 'custom':
            weight_ratio = trial.suggest_float('weight_ratio', 10, 500, log=True)
            common_params['class_weight'] = {0: 1, 1: weight_ratio}
        if solver in sklearn_solvers:
            params = {
                **common_params,
                'penalty': 'l2',
                'solver': solver
            }

            try:
                model = LogisticRegression(**params)
                model.fit(Xf_train, yu_train)
                y_pred = model.predict(Xf_val)
                y_prob = model.predict_proba(Xf_val)[:, 1] if hasattr(model, 'predict_proba') else None
                score, recall, precision, f_beta, pr_auc = custom_fraud_score(yu_val, y_pred, y_prob)
                trial.set_user_attr('recall', recall)
                trial.set_user_attr('precision', precision)
                trial.set_user_attr('f_beta', f_beta)
                if pr_auc is not None:
                    trial.set_user_attr('pr_auc', pr_auc)

                return score
            except Exception as e:
                return float('-inf')
        else:
            custom_params = {
                **common_params,
                'solver': solver
            }
            if solver == 'bfgs':
                custom_params['line_search'] = trial.suggest_categorical('line_search', ['armijo', 'wolfe'])
            elif solver == 'fixed-hessian':
                custom_params['hessian_update_freq'] = trial.suggest_int('hessian_update_freq', 5, 20)
            try:
                model = CustomLogisticRegression(**custom_params)
                model.fit(Xf_train, yu_train)

                y_pred = model.predict(Xf_val)
                y_prob = model.predict_proba(Xf_val)[:, 1] if hasattr(model, 'predict_proba') else None

                score, recall, precision, f_beta, pr_auc = custom_fraud_score(yu_val, y_pred, y_prob)

                trial.set_user_attr('recall', recall)
                trial.set_user_attr('precision', precision)
                trial.set_user_attr('f_beta', f_beta)
                if pr_auc is not None:
                    trial.set_user_attr('pr_auc', pr_auc)

                return score
            except Exception as e:
                return float('-inf')

    study = optuna.create_study(direction='maximize')
    study.optimize(objective, n_trials=50)

    best_params = study.best_params
    best_trial = study.best_trial


    if solver in sklearn_solvers:
        best_model_params = {
            'C': best_params['C'],
            'tol': best_params['tol'],
            'max_iter': best_params['max_iter'],
            'solver': solver,
            'penalty': 'l2',
            'random_state': 42
        }
        if 'class_weight_option' in best_params:
            if best_params['class_weight_option'] == 'balanced':
                best_model_params['class_weight'] = 'balanced'
            elif best_params['class_weight_option'] == 'custom' and 'weight_ratio' in best_params:
                best_model_params['class_weight'] = {0: 1, 1: best_params['weight_ratio']}

        best_model = LogisticRegression(**best_model_params)
    else:
        custom_model_params = {
            'C': best_params['C'],
            'max_iter': best_params['max_iter'],
            'solver': solver,
        }

        best_model = CustomLogisticRegression(**custom_model_params)

    best_model.fit(Xf_train, yu_train)

    best_models[solver] = best_model
    best_scores[solver] = best_trial.value

print("\n=== Test Set Evaluation ===")
test_metrics = {}

for solver, model in best_models.items():
    y_pred = model.predict(Xf_test_new)
    print(f"\nModel with solver: {solver}")
    print(classification_report(y_test_new, y_pred))

    plot_cm(y_test_new, y_pred, title=f"Confusion Matrix - {solver}")
```

We can see that because of the blackbox nature of the hyperparameters, randomized search doesn't produce results  much better (some are even slightly worse) than the previously observed ones, though it is worth noting that the scoring in the randomized search is not based on the scores viewed in the classification report.

## Undersampling & tuning of the probability threshold hyperparameter (1D grid search)

```python
from sklearn.metrics import f1_score

best_thresholds = {}
best_scores = {}

for model in models:
    print(f"Evaluating model: {model.solver}")
    model.fit(Xf_train, yu_train)
    best_score = 0.0
    best_threshold = 0.0

    for p_thresh in np.linspace(0.5, 0.95, 46):
        p_pred_proba = model.predict_proba(Xf_val)[:, 1]
        y_pred = [1 if p > p_thresh else 0 for p in p_pred_proba]
        current_score = f1_score(y_val, y_pred)
        if current_score > best_score:
            best_score = current_score
            best_threshold = p_thresh
    best_thresholds[model.solver] = best_threshold
    best_scores[model.solver] = best_score

print("\n=== Test Set Evaluation ===")
test_metrics = {}

for solver, model in best_models.items():
    y_pred = model.predict(Xf_test_new)
    print(f"\nModel with solver: {solver}")
    print(classification_report(y_test_new, y_pred))

    plot_cm(y_test_new, y_pred, title=f"Confusion Matrix - {solver}")
```

Again, we don't seem to notice any relevant improvement of the model's performance, meaning that the 5 models (with 5 different solvers) did not have to undergo parameter tuning.

# Summary - discussion and comparison of the solvers

Because the hyperparameter optimization didn't produce any significant improvements, in this summary we will focus on the models trained on the:

(a) Original dataset

(b) Undersampled dataset

(c) Undersampled dataset with reduced features

(a) **Original dataset**

We notice that training the models on the original dataset produced the highest precision overall, however these models also had the lowest recall, which results in the highest number of false negatives - which we want to minimize. We also notice that the complex and time-consuming models aren't that much better than the simpler models.

(b) **Undersampled dataset**

Undersampling definitely helped with the recall problem that we encountered. This also produced a very interesting result: the Fixed-Hessian (which is the least complex algorithm) actually had the highest precision and the highest F1-score - it only produced a couple more false negatives

(c) **Undersampled dataset with reduced features**

Despite producing approximately the same results as in (b), this approach is way more memory-efficient because it eliminated about a third of all the features, hence the memory used for the data is approximately 2/3 of the memory needed in (b). We notice again that the Fixed-Hessian method produces the highest precision and F1-score, but the lowest recall.

**Conclusion**

Based on qualitative observation, we find that simpler methods like Newton-Cholesky and Fixed-Hessian often perform comparably to more complex methods such as BFGS in our setting. In particular, Fixed-Hessian yields higher precision, though at the cost of lower recall â€” highlighting a tradeoff that may be desirable depending on the application. Thus, simpler methods may be more effective when computational efficiency or specific evaluation metrics are prioritized.
